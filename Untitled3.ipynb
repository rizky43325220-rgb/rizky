```python
class Student:
    def __init__(self, name, score):
        self.__name = name
        self.__score = 0
        self.__grade = "E"
        self.score = score  # pakai setter

    # Getter name
    @property
    def name(self):
        return self.__name

    # Setter name
    @name.setter
    def name(self, value):
        if value:
            self.__name = value
        else:
            print("Nama tidak boleh kosong")

    # Getter score
    @property
    def score(self):
        return self.__score

    # Setter score + update grade otomatis
    @score.setter
    def score(self, value):
        if 0 <= value <= 100:
            self.__score = value
            self.__update_grade()
        else:
            print("Nilai harus 0 - 100")

    # Method internal update grade
    def __update_grade(self):
        if self.__score >= 90:
            self.__grade = "A"
        elif self.__score >= 80:
            self.__grade = "B"
        elif self.__score >= 70:
            self.__grade = "C"
        elif self.__score >= 60:
            self.__grade = "D"
        else:
            self.__grade = "E"

    # Menampilkan info
    def show_info(self):
        print(f"Nama Mahasiswa: {self.__name}")
        print(f"Nilai: {self.__score}")
        print(f"Grade: {self.__grade}")

    # Destruktor
    def __del__(self):
        print(f"Data mahasiswa {self.__name} telah dihapus dari sistem.")


# MAIN PROGRAM
student1 = Student("Siti", 87)
student1.show_info()

print("\nNilai diubah...\n")
student1.score = 93
student1.show_info()

del student1
```
